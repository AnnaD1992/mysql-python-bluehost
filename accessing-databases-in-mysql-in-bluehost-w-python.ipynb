# 2. Accessing MySQL database from Bluehost with Python

It was really frustrating for me to access MySql with Python, so I want to share some of the knowledge I learned from my struggles. 

**My goal was:** To access Wordpress website's database (MySql), hosted on Bluehost

   1. **First of all, you should make your IP address recognizable when you access the database.** 
   
       - Bluehost --> Advanced --> below Databases-->remote MySql -->Add Access Host

           And at "Host" field you just put the IP of your computer. (check your IP through System Preferences of your computer)
           
           
   2. Second of all, you need to <strong>add a new user with a new password to the MySQL database </strong>, which you will use for connection through Python.
       - Bluehost --> Advanced --> below Databases -->  MySQL Databases 
*            There you add a new user and password
*            Then DO NOT FORGET To "Add User to the Database" . 
* You select your new user and add to the existing database and give ALL PRIVILIGES
           
 I did not do it in the beginning and my access to the database was always refused. 

```python
Then we go to the Jupyter Lab notebook and write magic
```

```python
# Define the MySQL engine using MySQL Connector/Python

engine = sqlalchemy.create_engine(
    'mysql+mysqlconnector://NEW_USER_NAME:PASSWORD@HOST:PORT/database_name'
).connect()

print(engine)
```

* NEW_USER_NAME - your new name you created in MySQL Databases, which you also added to the database
* PASSWORD - Your Password you set for your new user
* HOST - I tried "localhost", which should work by default, but it did not work for me, so I used the domain name of my website
* PORT - by default is 3306
* database_name - a precise name of your database (without .db at the end)
